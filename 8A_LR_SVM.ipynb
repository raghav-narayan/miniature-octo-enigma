```python
def draw_line(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    xmin = ((-coef[1]*mi - intercept)/coef[0]) 
    xmax = ((-coef[1]*ma - intercept)/coef[0])
    points=np.array([[xmin, mi],[xmax, ma]])
    plt.plot(points[:,0], points[:,1])
```

```python
#you can start writing code here.
#Fitting SVM on the datasets with different regularization strengths
 
C=[0.0001,1,100]
pos=0
for j,i in enumerate(ratios):  
    X_p=np.random.normal(0,0.05,size=(i[0],2))
    X_n=np.random.normal(0.13,0.02,size=(i[1],2))
    y_p=np.array([1]*i[0]).reshape(-1,1)
    y_n=np.array([0]*i[1]).reshape(-1,1)
    X=np.vstack((X_p,X_n))
    y=np.vstack((y_p,y_n))
    for c in C:
        plt.subplot(4, 3,pos+1)
        pos+=1
        plt.scatter(X_p[:,0],X_p[:,1])
        plt.scatter(X_n[:,0],X_n[:,1],color='red')
        clf =SVC(C=c,kernel='linear')
        clf.fit(X,y)
        draw_line(clf.coef_[0],clf.intercept_,np.min(X[:,1]),np.max(X[:,1]))   
        plt.title('Dataset'+str(j+1)+ ' C:'+str(c),fontsize=15)
plt.tight_layout()
plt.show()


```

```python
#Fitting the Logistic Regression for various regularization strengths
plt.figure(figsize=(20,15))
C=[0.0001,1,100]
pos=0
for j,i in enumerate(ratios):  
    X_p=np.random.normal(0,0.05,size=(i[0],2))
    X_n=np.random.normal(0.13,0.02,size=(i[1],2))
    y_p=np.array([1]*i[0]).reshape(-1,1)
    y_n=np.array([0]*i[1]).reshape(-1,1)
    X=np.vstack((X_p,X_n))
    y=np.vstack((y_p,y_n))
    for c in C:
        plt.subplot(4, 3,pos+1)
        pos+=1
        plt.scatter(X_p[:,0],X_p[:,1])
        plt.scatter(X_n[:,0],X_n[:,1],color='red')
        clf =LogisticRegression(C=c)
        clf.fit(X,y)
        draw_line(clf.coef_[0],clf.intercept_,np.min(X[:,1]),np.max(X[:,1]))   
        plt.title('Dataset'+str(j+1)+ ' C:'+str(c),fontsize=15)
plt.tight_layout()
plt.show()

```

<b>Observation(s)</b>:
<p>The optimization objective of SVM focuses on the tradeoff between margin maximization and the penalty to the misclassfication rate.</p>
Here the C - the hyperparameter for SVM gives the weightage to the penalties for the misclassification rate.
If the point is correctly classified, the penalty = 0, for incorrectly classified points, penalty != 0.
<p>When C=0.0001 , the margin maximization is given the higher priority, than the misclassification errors, so consequntly, however the dataset, balanced or imbalanced, the decision boundary is much further away from the data points. Highly underfit</p>
<p> When C = 1, SVM gives little weightage to the misclassification error. However even for imbalanced dataset, SVM does not classifiy correctly, but the model is much closer to the data points.</p>
For balanced datasets, the model is correctly classifiying most of the points, with the least errors.
<p> When C = 100, margin maximization is given the least priority than the misclassification. Due to large C value,  for imbalanced dataset,even though model is not classifying points correctly , model is the very closer to the data points.
As data becomes more balanced, the misclassifcation rate greatly reduces, and model is classifying most of the points correctly. However in this case, the margin is the shortest.</p>
