```python
# you should not import any other packages
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from sklearn.linear_model import SGDRegressor
```

```python
import numpy as np
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        res = sp.optimize.root(
            lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    return angles
```

```python
a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
X= b * np.sin(phi)
Y= a * np.cos(phi)
```

```python
outliers=[(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
alpha=[0.0001, 1, 100] 
fig=plt.figure(figsize=(20,15))
pos = 0
for i,a in enumerate(alpha):
    feature = X
    target = Y
    for j,outlier in enumerate(outliers): 
        ax=fig.add_subplot(3, 5,pos+1)
        pos+=1
        feature=np.append(feature,outlier[0])
        target=np.append(target,outlier[1])
        feature=feature.reshape(-1,1)
        target=target.reshape(-1,1)
        reg = SGDRegressor(alpha=a, eta0=0.001, learning_rate='constant',random_state=0)
        reg.fit(feature,target)
        plt.scatter(feature, target)
        abline(reg.coef_,reg.intercept_)
        plt.title( ' Alpha:'+str(a),fontsize=10)
        
plt.show()

```

```python
# Reference: https://stackoverflow.com/questions/7941226/how-to-add-line-based-on-slope-and-intercept-in-matplotlib
def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--',color='red')
```

<b>Observations</b>:
<p> Here the alpha gives the penalty to the model complexity(coeffecient of squared term for penalty) . Greater alpha tends to greater penalty.</p>
<p> When alpha = 0.0001, there is minimal weightage given for the penalty of badly predicted points(outliers), and the slope of the line can be easily influenced by outliers. The regression line is highly prone to overfitting. The weights of the optimization objective (OLS) dominates the loss function much than the regulizer term.  </p>
<p> When alpha = 1, there is some weightage given to penalty of the weights.As the no of outliers increase, insignificant weights are made lower . The slope of the line is more robust to changes in outliers than the previous case.
<p> When alpha = 100, slope is more rigid and robust to outliers. Reg term is given higher weightage, huge error points or outliers and their corresponding weights are made sparse if L1 regularizer, or heavy penalty if L2 regularizer. Model complexity is  made to be low. But if alpha is too high , all the weights could become insignificant leading to higher chances of underfitting.
